```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("world_population.csv")
df.head(10)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.rename(columns={"Area (km²)": "Area", "Density (per km²)": "Density","Country/Territory":"Country"}, inplace =True)
```

```python
df.columns
```

```python
sorted_df = df.sort_values(by=['Rank'])
```

```python
sorted_df
```

```python
top_10_countries = df.sort_values(by=['2022 Population'], ascending=False).head(10)
top_10_countries
```

```python
sns.barplot(x = "2022 Population", y = 'Country', data=top_10_countries)
plt.title("Top 10 countries by population in 2022")
plt.xlabel("Countries")
plt.ylabel("Population")
plt.show()
```

```python
sns.histplot(df["2022 Population"], bins=10, kde = True)
plt.title("Distribution of 2022 Population")
plt.xlabel("Population")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby('Continent')['2022 Population'].mean().sort_values(ascending=False).head(2)
```

```python
sns.scatterplot(x = "2022 Population", y = "Density", data = df)
plt.title("Population distribution of 2022")
plt.xlabel("Population")
plt.ylabel("Density")
plt.show()
```

```python
df.groupby('Continent')['2022 Population'].sum().sort_values(ascending=False)
```

```python
sns.violinplot(x="Continent", y = "Density", data = df, palette="viridis")
plt.title("Population density distribution across continent")
plt.xlabel("Continents")
plt.ylabel("Population Density")
plt.show()
```

```python
sns.boxplot(x="Continent", y = "Density", data = df, palette = "muted", showfliers = False)
plt.title("Population density distribution across continent (without outliers")
plt.xlabel("Continents")
plt.ylabel("Population Density")
plt.show()
```

```python
df['Population growth(2000-2022)'] = df['2022 Population'] - df['2000 Population']
```

```python
df.head()
```

```python
top_10_countries_2000_2022 = df[['Country','Population growth(2000-2022)']].sort_values(by='Population growth(2000-2022)', ascending = False).head(10)
```

```python
top_10_countries_2000_2022
```

```python
sns.barplot(x = "Population growth(2000-2022)", y = 'Country', data=top_10_countries_2000_2022)
plt.title("Poplation growth from 2000 to 2022")
plt.xlabel("Countries")
plt.ylabel("Population")
plt.show()
```

```python
# Average population growth in all the countries
Average_growth_rate = df['Growth Rate'].mean()
Average_growth_rate
```

```python
Average_density_continent = df.groupby('Continent')['Density'].mean().sort_values(ascending = False)
Average_density_continent = Average_density_continent.to_frame()
Average_density_continent
```

```python
sns.lineplot(x="Continent", y = "Density", data = Average_density_continent)
plt.title('Average density by Population')
plt.xlabel('Continent')
plt.ylabel('Average Density')
plt.show()
```

```python
total_population_2022 = df.groupby('Continent')['2022 Population'].sum().sort_values(ascending = False)
total_population_2022 = total_population_2022.to_frame()
total_population_2022
```

```python
sns.barplot(x = 'Continent', y = '2022 Population', data = total_population_2022)
plt.title('Total Population of 2022 by Continent')
plt.xlabel('Continent')
plt.ylabel('Population 2022')
plt.show()
```

```python
df['Population(1970-2022'] = df['1970 Population'] - df['2022 Population']

declined_df = df[df['Population(1970-2022']>0].sort_values(by = 'Population(1970-2022',ascending=False).head(5)
declined_df
```

```python
sns.barplot(x='Country', y = 'Population(1970-2022', data = declined_df)
plt.title('Country Population declined from 1970 to 2022')
plt.xlabel('Country')
plt.ylabel('Declined Population')
plt.show()
```
